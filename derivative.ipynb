посчитайте значение 1-ой и 2-ой производной фунции
f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1

```python
# Создаём функцию
import sympy as sy
sy.init_printing()
x = sy.Symbol('x')

def f(x):
    return x**5 + 4 * sy.sin(2*x) + sy.cos(3*x + 3)
```

```python
# Вот она
f(x)
```

```python
# Находим производную
def df(x):
    return sy.diff(f(x), x)

df(x)
```

```python
# Задаём значение x = 1
df(x).subs(x, 1)
```

```python
# Для проверки делаем то же самое методом derivative библиотеки scipy
from numpy import sin, cos
from scipy.misc import derivative

def f_(y):
    return y**5 + 4 * sin(2*y) + cos(3*y + 3)
# Кстати, вопрос, почему, если использовать не np.sin/cos(), а math.sin/cos(), у меня получается другой результат?
```

```python
derivative(f_, 1.0, n = 1, dx=1e-6)
```

```python
# Теперь производная второго порядка
def so(x):
    return sy.diff(f(x), x, x)

so(x)
```

```python
# Здесь всё аналогично :)
so(x).subs(x, 1)
```

Постройте график первой и второй производной функции
(sin(2x+1))^5
на отрезке [-5; 5]

```python
from scipy.misc import derivative
import numpy as np
import matplotlib.pyplot as plt

def g(x):
    return (sin(2*x + 1))**5

x = np.linspace(-5, 5)

gx = g(x)
g1x = [derivative(g, xx, dx=1e-6, n=1) for xx in x]
g2x = [derivative(g, xx, dx=1e-6, n=2) for xx in x]
```

```python
plt.plot(x, g(x), label='function')
plt.plot(x, g1x, label='derivative_first_order')
plt.plot(x, g2x, label='derivative_second_order')
plt.legend()
plt.show()
```
